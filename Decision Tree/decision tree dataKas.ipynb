```python
import pandas as pd
```

```python
df = pd.read_csv("dataKas.csv")
df.head()
```

```python
inputs = df.drop('Lunas atau Belum',axis='columns')
```

```python
target = df['Lunas atau Belum']
```

```python

from sklearn.preprocessing import LabelEncoder
le_Nama = LabelEncoder()
le_NPM = LabelEncoder()
le_Departemen = LabelEncoder()
```

```python
inputs['Nama_n'] = le_Nama.fit_transform(inputs['Nama'])
inputs['NPM_n'] = le_NPM.fit_transform(inputs['NPM'])
inputs['Departemen_n'] = le_Departemen.fit_transform(inputs['Departemen'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Nama','NPM','Departemen'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python

model.score(inputs_n,target)
```

```python
model.predict([[2,1,0]])
```
